```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
ds= pd.read_csv('tech_gadgets_rating_dataset.csv')
ds
```

```python
x=ds[['Price', 'Battery_Life', 'RAM_GB', 'Brand']]
y=ds[['Rating']]
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe= OneHotEncoder(sparse_output=False)
endata= ohe.fit_transform(x[['Brand']])
endata

column_name=ohe.get_feature_names_out(['Brand'])
endf= pd.DataFrame(endata , columns=column_name)
endf

x= pd.concat([x,endf], axis=1)

x= x.drop(columns=['Brand'])
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(random_state=1)
reg.fit(x_train,y_train)
```

```python
y_pred= reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
score= r2_score(y_test,y_pred)
score
```
